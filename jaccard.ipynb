```python
import re
```

```python
# preprocessing data
def solution(str1, str2) -> int:
    new_str1 = ""
    new_str2 = ""

    for a in str1:
        if re.match('[a-zA-Z]', a):
            new_str1 += a

    for a in str2:
        if re.match('[a-zA-Z]', a):
            new_str2 += a
            
    # make sets of strs
    multi_value_set_1 = []
    multi_value_set_2 = []
    for n in range(len(new_str1) - 1):
        multi_value_set_1.append(new_str1[n:n+2].lower())

    for n in range(len(new_str2) - 1):
        multi_value_set_2.append(new_str2[n:n+2].lower())    

    # make a intersection set and an union set
    inter = []
    uni = []

    list2set_1 = set(multi_value_set_1)
    list2set_2 = set(multi_value_set_2)

    for ele in list2set_1:
        if ele in list2set_2:
            ele_count1 = multi_value_set_1.count(ele)
            ele_count2 = multi_value_set_2.count(ele)
            if ele_count1 <= ele_count2:
                # intersection 
                i = 1
                while i <= ele_count1:
                    inter.append(ele)
                    i += 1
                    if i > ele_count1:
                        break
                # union
                j = 1
                while j <= ele_count2:
                    uni.append(ele)
                    j += 1
                    if j > ele_count2:
                        break
            else:
                # intersection 
                i = 1
                while i <= ele_count2:
                    inter.append(ele)
                    i += 1
                    if i > ele_count2:
                        break
                # union
                j = 1
                while j <= ele_count1:
                    uni.append(ele)
                    j += 1
                    if j > ele_count1:
                        break
        else:
            # union
            ele_count3 = multi_value_set_1.count(ele)
            l = 1
            while l <= ele_count3:
                uni.append(ele)
                l += 1
                if l > ele_count3:
                    break
    # union
    for ele in list2set_2:
        if ele not in list2set_1:
            ele_count4 = multi_value_set_2.count(ele)
            m = 1
            while m <= ele_count4:
                uni.append(ele)
                m += 1
                if m > ele_count4:
                    break
                
    # result
    try:
        jaccard = int(len(inter) / len(uni) * 65536)
    except ZeroDivisionError:
        jaccard = 1
    return jaccard
```
